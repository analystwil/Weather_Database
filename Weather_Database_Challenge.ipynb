```python
#!pip install citipy
from citipy import citipy
import numpy as np
import pandas as pd
import time
import requests
from datetime import datetime
from pprint import pprint

# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

```python
coordinate = coordinates[0]
print(coordinates[0])
nearest_city_data = citipy.nearest_city(coordinate[0], coordinate[1])
print(nearest_city_data)
city_name = nearest_city_data.city_name
country_code = nearest_city_data.country_code
```

```python
# Create a list for holding the cities.
cities = []

# Identify nearest city for each of the 2,000 latitude & longitude combinations
for coordinate in coordinates:
    
    # Pull nearest city using lat/lon combo
    nearest_city_data = citipy.nearest_city(coordinate[0], coordinate[1])
    city_name = nearest_city_data.city_name
    country_code = nearest_city_data.country_code
    
    # If the city is unique, then add it to cities list
    if city_name not in cities:
        print(f"{city_name} has a country code of {country_code}")
        cities.append(city_name)

# Print the city count to confirm sufficient count
len(cities)
```
